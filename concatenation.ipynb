```python
import pandas as pd

path = "data/cleaned/"
articles = pd.read_csv(path + 'articles.csv', index_col="Unnamed: 0")
life_situations = pd.read_csv(path + 'life_situations.csv', index_col="Unnamed: 0")
news = pd.read_csv(path + 'news.csv', index_col="Unnamed: 0")
services = pd.read_csv(path + 'services.csv', index_col="Unnamed: 0")
```

```python
def concat_text_columns(df, columns):
    df = df.copy()
    df['text'] = df[columns].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
    df['text'] = df['text'].str.replace('nan ', '').str.replace(' nan ', '').str.replace(' nan', '')
    df = df.drop(columns=columns)
    return df
```

```python
articles_concat = concat_text_columns(articles, ['title', 'content'])
articles_concat
```

```python
life_situations_concat = concat_text_columns(life_situations, ['intro', 'title_main', 'title_sub'])
life_situations_concat
```

```python
news_concat = concat_text_columns(news, ['title', 'body', 'short_description'])
news_concat
```

```python
services_concat = concat_text_columns(services, ['additional_info', 'description', 'full_title', 'result_description'])
services_concat = services_concat.drop(columns=['short_title', 'title'])
services_concat
```

```python
# save it into concated folder
path = "data/concated/"
articles_concat.to_csv(path + 'articles.csv')
life_situations_concat.to_csv(path + 'life_situations.csv')
news_concat.to_csv(path + 'news.csv')
services_concat.to_csv(path + 'services.csv')
```
